# READING DATA


```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("iris.csv")
```

```python
data.head()
```

```python
data.describe(include="all")
```

#### The data consist of 150 records.

#### The above table gives some basic statistical details of the data.

```python
data=data.drop("Id",axis=1)
```

# IMPORTING VISUALIZATION LIBRARIES


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(data , hue='Species')
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

# SPLITTING THE DATA INTO TRAINING AND TESTING DATA


```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

# IMPORTING DECISION TREE LIBRARIES


#### DECISION TREE

Decision Trees are a type of Supervised Machine Learning where the data is continuously split according to a certain parameter.


#### ENTROPY

Entropy can be defined as a measure of the purity of the sub split.


#### INFORMATION GAIN

Information gain is the reduction in entropy or surprise by transforming a dataset and is often used in training decision trees. Information gain is calculated by comparing the entropy of the dataset before and after a transformation.



#### GINI IMPURITY

The Gini impurity measure is one of the methods used in decision tree algorithms to decide the optimal split from a root node, and subsequent splits.Note that the lower the Gini the better the split.



#### GINI IMPURITY VS ENTROPY


```python
from sklearn import tree
dtc=tree.DecisionTreeClassifier(random_state=0)
dtc.fit(X_train,y_train)
```

```python
y_pred1=dtc.predict(X_train)
from sklearn.metrics import accuracy_score
accuracy_score(y_train,y_pred1)
```

```python
y_pred=dtc.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# VISUALIZING THE TREE

```python
plt.figure(figsize=(15,10))
tree.plot_tree(dtc,filled=True)
```

## we feed new data to this classifier in the form of test data and precision was 97%.
